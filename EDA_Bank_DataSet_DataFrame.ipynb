```python
job_list = bank_df['job'].unique()
job_list
```

```python
job_count_list=[]
for i in bank_df['job'].unique():
    con=bank_df['job']==i
    count=len(bank_df[con])
    job_count_list.append(count)

job_count_list
```

```python
data = zip(job_list, job_count_list)
cols=['Job', 'Count']
job_data_df = pd.DataFrame(data, columns=cols)
job_data_df
```

```python
plt.bar('Job', 'Count', data=job_data_df)
plt.title('Job Count Bar Chart')
plt.xlabel('Job')
plt.ylabel('Count')
plt.savefig('Job_Count_Bar_Chart.jpg')
plt.show()
```

```python
marital_list = bank_df['marital'].unique()
marital_count_list=[]

for i in bank_df['marital'].unique():
    marital_count_list.append(len(bank_df[bank_df['marital']==i]))

print(marital_list)
print(marital_count_list)
```

```python
data=zip(marital_list,marital_count_list)
cols=['Merit Status','Merit Count']
marital_data_df = pd.DataFrame(data, columns=cols)
print(marital_data_df)
plt.bar('Merit Status','Merit Count',data=marital_data_df)
plt.title('Merit Bar Chart')
plt.xlabel('Merit Status')
plt.ylabel('Merit Count')
plt.savefig('Merit_Bar_Chart.jpg')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

def bank_chart(i):
    plt.figure(figsize=(8,5))
    order_continents = bank_df[i].value_counts().keys()
    sns.countplot(data=bank_df,
                 x=i,
                 order=order_continents)
    plt.title(f'Bank {i} Chart Using Seaborn')
    plt.savefig(f"{bank_path}\\{i}_chart_using_seaborn.jpg")
    plt.show()

cat_columns = bank_df.select_dtypes(include='object').columns

for key in cat_columns:
    bank_chart(key)

```
