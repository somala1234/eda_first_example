```python
**** Empirical rule:  ********
mean-1std to mean+1std : 68%
mean-2std to mean+2std : 95%
mean-3std to mean+3std : 99.7%
```

```python
https://builtin.com/sites/www.builtin.com/files/styles/ckeditor_optimize/public/inline-images/1_empirical-rule.jpg
```

```python
Step-1 : Calculate the mean of prevailining wage
Step-2 : Calculate std of prevailing wage
Step-3 : lb = mean-1 * std
Step-4 : up = mean+1 * std
Step-5 : con1 = data < ub
Step-6 : con2= data > lb
Step-7 : con3 = con1 and con2
Step-8 : data[con3]
Step-9 : compare with 68% 25480 count
```

```python
wage_data = visa_df['prevailing_wage']
mu = visa_df['prevailing_wage'].mean()
sigma = visa_df['prevailing_wage'].std()
lb = mu-sigma
ub = mu + sigma
con1=wage_data<ub
con2=wage_data>lb
con3=con1 & con2
wage_68p=wage_data[con3]
print(wage_68p)
print(68*len(visa_df)/100)
```

```python
**************95%********************
Step-1 : Calculate the mean of prevailining wage
Step-2 : Calculate std of prevailing wage
Step-3 : lb = mean-2 * std
Step-4 : up = mean+2 * std
Step-5 : con1 = data < ub
Step-6 : con2= data > lb
Step-7 : con3 = con1 and con2
Step-8 : data[con3]
Step-9 : compare with 95% 25480 count
```

```python
wage_data = visa_df['prevailing_wage']
mu = visa_df['prevailing_wage'].mean()
sigma = visa_df['prevailing_wage'].std()
lb = mu-2*sigma
ub = mu+2*sigma
con1=wage_data<ub
con2=wage_data>lb
con3=con1 & con2
wage_95p=wage_data[con3]
print(wage_95p)
print(95*len(visa_df)/100)
```

```python
**************98%********************
Step-1 : Calculate the mean of prevailining wage
Step-2 : Calculate std of prevailing wage
Step-3 : lb = mean-3 * std
Step-4 : up = mean+3 * std
Step-5 : con1 = data < ub
Step-6 : con2= data > lb
Step-7 : con3 = con1 and con2
Step-8 : data[con3]
Step-9 : compare with 93% 25480 count
```

```python
wage_data = visa_df['prevailing_wage']
mu = visa_df['prevailing_wage'].mean()
sigma = visa_df['prevailing_wage'].std()
lb = mu-3*sigma
ub = mu+3*sigma
con1=wage_data<ub
con2=wage_data>lb
con3=con1 & con2
wage_98p=wage_data[con3]
print(wage_98p)
print(98*len(visa_df)/100)
```

```python
*** How to deal outliers ****
    -1. We can drop outliers
        -- Dropping outliers is an easy task, but not recomonded
        -- If you drop the outliers, we are dropping the rows
        -- Which means we are dropping other caolumn values aslo
        -- May be some important information we might loose from other column values
        -- But in industry if outlier percentage less than 2% of the total data , we can drop.
          -- The number of outliers in wage is 427
         -- The number of total observations are 25480
        -- The percentage is 1.67 which is less than 2%

    -2. Fill with median value
        -- We already know that outliers does not effect by median
        -- So it is the best practive to fill the outliers with median

    -3. Capping or Florring
        - Instead of fill with mdian
            -we can certain values
            -The outliers which are less than LB fill with LB vlaue
            - The outliers which are greater than UB, fill with UB value
   

    ******Isolation Forest******
    - Is Model used to find the outliers and anamolies
    
  
```
